```python
import pickle
```

```python
# Load saved model
with open("fake_news_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)

# Load saved TF-IDF vectorizer
with open("tfidf_vectorizer.pkl", "rb") as f:
    loaded_tfidf = pickle.load(f)

print("Model and vectorizer loaded successfully")
```

```python
import re

def clean_text_simple(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
def predict_news_loaded(news_text):
    cleaned = clean_text_simple(news_text)
    vector = loaded_tfidf.transform([cleaned])
    prediction = loaded_model.predict(vector)
    
    return "REAL NEWS ðŸŸ¢" if prediction[0] == 1 else "FAKE NEWS ðŸ”´"
```

```python
sample_news = """
Breaking News: Scientists confirm a new renewable energy source
that could significantly reduce carbon emissions worldwide.
"""

print(predict_news_loaded(sample_news))
```

```python
sample_news = """
WASHINGTON (Reuters) - The U.S. Senate voted on Tuesday to pass a new bill
focused on healthcare reforms proposed by the administration.
"""

print(predict_news_loaded(sample_news))
```
